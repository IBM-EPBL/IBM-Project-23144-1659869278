# **Outliers Replacement**

```python
sns.boxplot(df1.Age)
```

```python
q1=df1.Age.quantile(0.25)
q3=df1.Age.quantile(0.75)
```

```python
IQR = q3-q1
```

```python
upper_limit=q3 + 1.5 * IQR
```

```python
upper_limit
```

```python
df1.describe()
```

```python
df1.Age=np.where(df1.Age>upper_limit,44,df1.Age) #Median=44.000000	
```

```python
sns.boxplot(df1.Age)
```
