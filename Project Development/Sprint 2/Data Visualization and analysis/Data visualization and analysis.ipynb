## Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import sklearn
```

## Loading dataset

```python
ds=pd.read_csv("C:\loan_prediction.csv")
```

```python
ds.head()
```

## Univariate analysis

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
sb.distplot(ds['ApplicantIncome'],color='r')
plt.subplot(122)
sb.distplot(ds['Credit_History'])
plt.show()
```

```python
sb.histplot(ds['LoanAmount'])
```

## Bivariate analysis

```python
sb.countplot(x ='Education',data = ds)
```

```python
sb.countplot(x ='Gender',data = ds)
```

```python
sb.barplot(x='ApplicantIncome', y='LoanAmount',data=ds)
plt.show()
```

```python
plt.hist(x = ds.Loan_Status, bins = 3, color="brown")
plt.title('comparison of Loan Status with Income')
plt.xlabel('Loan_Status')
plt.ylabel('ApplicantIncome')
plt.show()
```

```python
sb.scatterplot(data=ds, x="LoanAmount", y="Loan_Amount_Term")
```

## Multivariate analysis

```python
sb.swarmplot(x="Gender",y="ApplicantIncome",hue="Loan_Status",data=ds)
```

```python
sb.swarmplot(x="Education",y="ApplicantIncome",hue="Loan_Status",data=ds)
```

```python
sb.pairplot(ds)
```
